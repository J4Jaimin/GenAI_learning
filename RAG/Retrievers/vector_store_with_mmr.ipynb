```python
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS
from langchain_core.documents import Document
```

```python
documents = [

    # 🔥 Same topic (FastAPI) — high redundancy
    {"id": 1, "text": "FastAPI is a fast Python web framework for building APIs.", "metadata": {"topic": "fastapi"}},
    
    {"id": 2, "text": "FastAPI allows developers to build high-performance APIs quickly using Python.", "metadata": {"topic": "fastapi"}},
    
    {"id": 3, "text": "FastAPI supports async programming and automatic API documentation.", "metadata": {"topic": "fastapi"}},

    {"id": 4, "text": "FastAPI is widely used for backend development due to its speed and ease of use.", "metadata": {"topic": "fastapi"}},


    # ⚡ Related but slightly different (Backend ecosystem)
    {"id": 5, "text": "Django is a Python web framework that follows the MVC pattern.", "metadata": {"topic": "django"}},
    
    {"id": 6, "text": "Flask is a lightweight Python web framework used for building simple APIs.", "metadata": {"topic": "flask"}},


    # 🚀 Completely different (Vector DB / AI)
    {"id": 7, "text": "Vector databases store embeddings for similarity search in AI applications.", "metadata": {"topic": "vector_db"}},
    
    {"id": 8, "text": "FAISS is a library developed by Facebook for efficient similarity search.", "metadata": {"topic": "faiss"}},

    {"id": 9, "text": "Embeddings convert text into numerical vectors for machine learning models.", "metadata": {"topic": "embeddings"}},


    # 🎯 Different domain (Noise for diversity testing)
    {"id": 10, "text": "Mumbai is one of the largest cities in India and a financial hub.", "metadata": {"topic": "general"}},
    
    {"id": 11, "text": "Cricket is the most popular sport in India with millions of fans.", "metadata": {"topic": "sports"}},

]
```

```python
embedding = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-mpnet-base-v2",
    model_kwargs={"device": "cpu"}
)
```

```python
vector_store = FAISS.from_documents(
    documents = [Document(page_content=doc["text"], metadata=doc["metadata"]) for doc in documents],
    embedding = embedding
)
```

```python
retriever = vector_store.as_retriever(
    search_type="mmr",
    search_kwargs={"k": 3, "lambda_mult": 0.5}
)
```

```python
results = retriever.invoke("What is the key features of FastAPI?")
```

```python
for result in results:
    print(result.metadata["topic"], ":", result.page_content)
```

```python
docs_dict = vector_store.docstore._dict

for doc_id, doc in docs_dict.items():
    print(doc_id, doc.page_content)
```
