```python
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_community.vectorstores import Chroma
from langchain_core.documents import Document
```

```python
documents = [
    {
        "id": 1,
        "text": "FastAPI is a modern web framework for building APIs with Python. It is known for its speed and automatic documentation using Swagger UI.",
        "metadata": {"topic": "fastapi"}
    },

    {
        "id": 2,
        "text": "Node.js is a JavaScript runtime built on Chrome's V8 engine. It is commonly used for building scalable backend applications.",
        "metadata": {"topic": "nodejs"}
    },

    {
        "id": 3,
        "text": "MongoDB is a NoSQL database that stores data in flexible JSON-like documents. It is widely used in MERN stack applications.",
        "metadata": {"topic": "mongodb"}
    },

    {
        "id": 4,
        "text": "Vector databases store embeddings and allow similarity search. FAISS and Chroma are popular tools for building vector stores.",
        "metadata": {"topic": "vector_db"}
    },

    {
        "id": 5,
        "text": "LangChain is a framework for building applications using large language models. It helps in chaining prompts, memory, and tools.",
        "metadata": {"topic": "langchain"}
    },

    {
        "id": 6,
        "text": "Retrievers are used in RAG systems to fetch relevant documents based on user queries. They use similarity search over embeddings.",
        "metadata": {"topic": "retriever"}
    },

    {
        "id": 7,
        "text": "Docker allows developers to package applications and their dependencies into containers, ensuring consistency across environments.",
        "metadata": {"topic": "docker"}
    },

    {
        "id": 8,
        "text": "Pandas is a Python library used for data manipulation and analysis. It provides data structures like DataFrames.",
        "metadata": {"topic": "pandas"}
    },

    {
        "id": 9,
        "text": "Embeddings convert text into numerical vectors. These vectors capture semantic meaning and are used in similarity search.",
        "metadata": {"topic": "embeddings"}
    },

    {
        "id": 10,
        "text": "Chroma is an open-source embedding database. It is commonly used with LangChain for storing and retrieving documents.",
        "metadata": {"topic": "chroma"}
    }

]
```

```python
embedding = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-mpnet-base-v2",
    model_kwargs={"device": "cpu"}
)
```

```python
vector_store = Chroma.from_documents(
    documents = [Document(page_content=doc["text"], metadata=doc["metadata"]) for doc in documents],
    embedding = embedding,
    collection_name = "my_collection"
)
```

```python
retriever = vector_store.as_retriever(search_kwargs={"k": 3})
```

```python
query = "What is the best vector database for langchain?"

results = retriever.invoke(query)
```

```python
for doc in results:
    print(f"Content: {doc.page_content}")
    print(f"Metadata: {doc.metadata}")
    print("-" * 50)
```
