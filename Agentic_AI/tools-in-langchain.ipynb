## Custom tool using pydantic schema and structured tool.

```python
from pydantic import BaseModel, Field
from typing import Type
```

```python
class MultiplyArgs(BaseModel):
    a: int = Field(required=True, description="The first number to multiply.")
    b: int = Field(required=True, description="The second number to multiply.")
```

```python
multiply_tool = StructuredTool.from_function(
    func = multiply,
    name = "multiply",
    description = "Multiplies two numbers.",
    args_schema = MultiplyArgs
)
```

```python
result = multiply_tool.invoke({"a": 4, "b": 6})

print(result)
```

```python
print(multiply_tool.name)
print(multiply_tool.description)
print(multiply_tool.args_schema.model_json_schema())
```

## Custom tool using a BaseTool

```python
class MultiplyToolUsingBaseTool(BaseTool):
    name: str = "multiply"
    description: str = "Multiplies two numbers."
    args_schema: Type[BaseModel] = MultiplyArgs

    def _run(self, a: int, b: int) -> int:
        """Multiplies two numbers."""
        return a * b
```

```python
multiply_tool_using_base_tool = MultiplyToolUsingBaseTool()
```

```python
multiply_tool_using_base_tool.invoke({"a": 7, "b": 8})
```

```python
print(multiply_tool_using_base_tool.name)
print(multiply_tool_using_base_tool.description)
print(multiply_tool_using_base_tool.args_schema.model_json_schema())
```

# Toolkit

```python
# custom tools

@tool 
def add(a: int, b: int) -> int:
    """Adds two numbers."""
    return a + b

@tool
def subtract(a: int, b: int) -> int:
    """Subtracts two numbers."""
    return a - b
```

```python
class MathToolkit:
    def get_tools(self):
        return [add, subtract]
```

```python
toolkit = MathToolkit()

tools = toolkit.get_tools()
```

```python
for tool in tools:
    print(tool.name)
    print(tool.description)
    print(tool.args_schema.model_json_schema())
```
