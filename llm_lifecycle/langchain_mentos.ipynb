```python
import random
from abc import ABC, abstractmethod
```

```python
class Runnable(ABC):

    @abstractmethod
    def invoke(self, input_data):
        pass
```

```python
class NakliLLM(Runnable):
    def __init__(self):
        print("LLM created")

    def invoke(self, prompt):
        
        response_list = [
            "Delhi is the capital of India.",
            "IPL is the most popular cricket league in the world.",
            "AI stands for Artificial Intelligence.",
            "Python is a popular programming language.",
        ]

        return {'response': random.choice(response_list)}

    def predict(self, prompt):

        response_list = [
            "Delhi is the capital of India.",
            "IPL is the most popular cricket league in the world.",
            "AI stands for Artificial Intelligence.",
            "Python is a popular programming language.",
        ]

        return {'response': random.choice(response_list)}
```

```python
llm = NakliLLM()
```

```python
class NakliPromptTemplate(Runnable):
    def __init__(self, template, input_variables):
        self.template = template
        self.input_variables = input_variables

        print("Prompt Template created")

    def invoke(self, input_dict):
        return self.template.format(**input_dict)

    def format(self, input_dict):
        return self.template.format(**input_dict)
```

```python
template = NakliPromptTemplate(
    template = "What is the capital of {country}?",
    input_variables = ["country"]
)
```

```python
class OutputParser(Runnable):
    def __init__(self):
        print("Output Parser created")

    def invoke(self, input_data):
        response = input_data['response']
        return response
```

```python
outputParser = OutputParser()
```

```python
class RunnableConnector(Runnable):
    def __init__(self, runnable_list):
        self.runnable_list = runnable_list

    def invoke(self, input_data):
        for runnable in self.runnable_list:
            input_data = runnable.invoke(input_data)

        return input_data
```

```python
chain = RunnableConnector([template, llm, outputParser])
```

```python
chain.invoke({"country": "India"})
```

## Merge two different chains with the help of Runnables.

```python
template1 = NakliPromptTemplate(
    template = "Write a joke about {topic}.",
    input_variables = ["topic"]
)
```

```python
template2 = NakliPromptTemplate(
    template = "Explain the joke: {response}",
    input_variables = ["response"]
)
```

```python
chain1 = RunnableConnector([template1, llm])
```

```python
chain2 = RunnableConnector([template2, llm, outputParser])
```

```python
final_chain = RunnableConnector([chain1, chain2])
```

```python
final_chain.invoke({"topic": "programming"})
```
