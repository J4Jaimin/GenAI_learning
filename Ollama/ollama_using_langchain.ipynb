```python
from langchain_ollama import ChatOllama
from langchain_ollama import OllamaLLM
from langchain_ollama import OllamaEmbeddings
```

```python
# chat ollama for chat with models running on ollama server

llm = ChatOllama(
    model = "llama3.2:3b",
    temperature = 0.3
)

response = llm.invoke("Explain the concept of Langchain in simple terms.")

print(response.content)
```

```python
# llm for text generation

llm = OllamaLLM(
    model = "llama3.2:3b",
    temperature = 0.3
)

response = llm.invoke("What is the capital of France?")

print(response)
```

```python
# embeddings generation

embeddings = OllamaEmbeddings(
    model = "embeddinggemma:latest"
)

query_result = embeddings.embed_query("India")

print(query_result)
```

```python
# Access ollama cloud models with ollama signin

llm = ChatOllama(
    model = "qwen3.5:397b-cloud",
    temperature = 0.3
)

response = llm.invoke("Explain the concept of Langchain in simple terms.")

print(response.content)
```
