```python
import ollama
from ollama import chat
```

```python
response = ollama.generate(model="tinyllama", prompt="Please give me a joke and its explanation.")
```

```python
print(response.response)
```

```python
response = ollama.generate(model="tinyllama", prompt="Why don't scientists trust atoms? give me in report format.", stream=True)

for chunk in response:
    print(chunk["response"], end="")
```

```python
result = ollama.generate(model="tinyllama", prompt="Why is the sky blue?", system="You are a funny assistance, reply in funny way.")

print(result.response)
```

```python
stream = chat(
    model='llama3.2:3b',
    messages=[{'role': 'user', 'content': 'Why is the sky blue?'}],
    stream=True,
)

for chunk in stream:
  print(chunk['message']['content'], end='', flush=True)
```

```python
model_details = ollama.show(model="llama3.2:3b")
```
